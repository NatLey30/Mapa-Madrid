```python
import networkx as nx
import matplotlib.pyplot as plt
```

GRAFOS NO DIRIGIDO

```python
G = nx.Graph()
G.add_node(1)
G.add_node(2)
G.add_edge(1,2)
G.add_nodes_from([3,4,5])
G.add_edges_from([(1,4),(3,4),(2,4),(3,2),(3,5)])
print(G)
print(G.nodes)
print(G.edges)
```

```python
G.remove_node(5)
print(G)
print(G.nodes)
print(G.edges)
```

```python
#Adyacencia
print(G[1])
print(G.adj[1])
```

```python
plot = plt.plot()
nx.draw(G, with_labels=True, pos={1:(0,0), 2:(0,1), 3:(1,0), 4:(1,1)})
plt.show
```

PROBLEMA 5

```python
G2 = nx.Graph()
G2.add_nodes_from(['a','b',10,11,3.14])
print(G2.nodes)

G2.add_edge('a',10, object='una cadena', weight=1.1)
print(G2['a'])

G2.add_edges_from([('a',3.14,{'object':'otra cadena', 'weight':1.2}), ('b', 11, {'object':'un objeto', 'weight':5.7})])
print(G2.edges)

print(G2.edges['a',10])
print(G2['b'][11])

plot = plt.plot()
nx.draw(G2, with_labels=True)
plt.show()
```

GRAFOS DIRIGIDOS

```python
Gd = nx.DiGraph()
Gd.add_nodes_from([1,2,3,4])
Gd.add_edges_from([(1,2),(2,1),(1,4),(3,4),(2,4),(4,2),(3,2)])
plot = plt.plot()
nx.draw(Gd, with_labels=True)
plt.show()
```

Code

```python
print(G.degree())
print(G.degree(1))
print(G2.degree())
print(Gd.in_degree())
print(Gd.out_degree())
```

```python
print(list(nx.connected_components(G2)))

print(list(nx.weakly_connected_components(Gd)))
print(list(nx.strongly_connected_components(Gd)))
```

```python
aam = nx.minimum_spanning_tree(G)
print(aam.nodes)
print(aam.edges)

plot = plt.plot()
pos = nx.spring_layout(G)
nx.draw(G, with_labels=True, pos=pos, edge_color='b')
nx.draw(aam ,with_labels=False, pos=pos, edge_color='r')
plt.show()
```

```python
# Camino ms corte entre 1 y 3
camino = nx.shortest_path(G,1,3)
print(camino)

plot = plt.plot()
pos = nx.spring_layout(G)
nx.draw(G, with_labels=True, pos=pos, edge_color='b')
nx.draw_networkx_edges(G, pos=pos, arrows=True, edgelist=[(1,4),(4,3)], edge_color='r')
nx.draw_networkx_edge_labels(G, pos=pos, edge_labels={(1,4):'1', (4,3): '2'})
plt.show()
```
